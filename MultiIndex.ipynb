```python
import pandas as pd
```

```python
import numpy as np
```

```python
# 创建一个含有分层索引的dataframe。先赋值0层索引，再赋值1、2..N层索引。
#多层索引中，每个轴上的一行或一列索引用level表示，level=0表示最外层的索引。索引上的标签用label表示。
df = pd.DataFrame(np.arange(16).reshape(4,4),index=[['a','a','b','b'],[1,2,1,2]],columns=[['A','A','B','B'],['Big','Small','Big','Small']])
```

```python
df
```

```python
#对行列索引进行命名。
df.index.names=['Char','Num']
df.columns.names=['BigChar','Size']
df
```

```python
#查看分层索引，返回一个多层索引对象。
df.index
```

```python
#分层索引如何选择数据。先索引0层，再索引1、2...N层。
df['A']
```

```python
#选择行索引
df.loc['a']
```

```python
#选择列索引下的索引
df['A','Big']
```

```python
#unstack，解除堆叠，将行索引转换为列索引，默认将最内层索引转换为列的最内层索引，传递数字0\1\N或者level=0\1\N，可指定将第N层索引转换为列的最内层索引。
df.unstack(0)
```

```python
#stack，堆叠，和unstack相反，将列索引堆叠为行的最后一层索引。默认是堆叠列最内层索引，传递level参数可以堆叠N层列索引。
df.stack(0)
```

```python
#已有索引，如何构建多层索引。
df = pd.DataFrame(np.arange(16).reshape(4,4),index=[3,4,5,6],columns=[7,8,9,10])
df.index = [['a','a','b','b'],[1,2,1,2]]
df.columns = [['A','A','B','B'],['Big','Small','Big','Small']]
df
```

```python
#swaplevel使用，可以使用数字标签level进行标签互换,传递axis=1可以互换列索引。此操作不修改原数据
df.swaplevel(1,0)
df.index.names=['Char','Num']
df.columns.names=['BigChar','Size']
df.swaplevel('Size','BigChar',axis=1)
```

```python
#sort_index可以对多层索引进行排序，但只能排一个索引，传递level即可，也接受axis参数。
df.sort_index(level=1)
df.sort_index(level=1,axis=1)
```
